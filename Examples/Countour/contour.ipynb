```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Find Coordinates of Contours

```python
# Load image
font = cv.FONT_HERSHEY_COMPLEX
image = cv.imread("../Images/shapes.jpg", cv.IMREAD_COLOR)

# Load as grayscale
gray = cv.imread("../Images/shapes.jpg", cv.IMREAD_GRAYSCALE)

# Converting image to a binary image
_, threshold = cv.threshold(gray, 110, 255, cv.THRESH_BINARY)

# Detecting contours in image.
contours, _ = cv.findContours(threshold, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

for contour in contours:
    approx = cv.approxPolyDP(contour, 0.009 * cv.arcLength(contour, True), True)

    # Draw boundary of contours
    cv.drawContours(image, [approx], 0, (0, 0, 255), 5) 

plt.axis("off")
plt.imshow(image)
plt.show()
```
