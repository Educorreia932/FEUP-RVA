# Homography

```python
import cv2 as cv
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_source = cv.imread("../Images/ebook1.jpg")
image_target = cv.imread("../Images/ebook2.jpg")

image_source = cv.cvtColor(image_source, cv.COLOR_BGR2RGB)
image_target = cv.cvtColor(image_target, cv.COLOR_BGR2RGB)

# Four corners of the e-book 
points_source = np.array([
    (490, 2770),
    (1790, 2760),
    (1705, 1025),
    (545, 1040)
])

points_target = np.array([
    (910, 2680),
    (2160, 2150),
    (1330, 825),
    (300, 1110)
])

h, status = cv.findHomography(points_source, points_target)

image_out = cv.warpPerspective(image_source, h, (image_target.shape[1], image_target.shape[0]))

images = (image_source, image_target, image_out)

figure = plt.figure(figsize=(10, 30))

for i, image in enumerate(images):
    figure.add_subplot(1, len(images), i + 1)
    plt.subplot(len(images) + 1, 3, i + 1)
    plt.axis("off")
    plt.imshow(image)
```
